```python
col_list=[]
for i in data.columns:
  if((data[i].dtype=='object' ) & (i != 'Churn')):
    col_list.append(i)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in col_list:
  data[i]=le.fit_transform(data[i])
```

```python
data.info()
```

```python
data.head()
```

```python
x=data.iloc[:,:-1]
y=data['Churn']
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
# build the model
classifier1= RandomForestClassifier(n_estimators=100)# 100 desion tree taken , how we can decide no. of dision tree
classifier1.fit(x_train,y_train)
```

```python
# now predict , what the model have learnt by PREDICT function
y_pred= classifier1.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)# just compare predict value and test value
```

```python
# how can decide optimal no of desion tree
# hit and trial method
# let make a list of ni of desion tree
li=[50,100,150,200,250,300,500,700,900]
accuracy_li=[]

for i in li:
  rd_model= RandomForestClassifier(n_estimators=i)
  rd_model.fit(x_train,y_train)
  pred= rd_model.predict(x_test)
  a=accuracy_score(y_test, pred)
  accuracy_li.append(a)
```

```python
accuracy_li
```
